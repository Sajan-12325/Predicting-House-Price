## Random Sample Consensus(RANSAC) Algorithm

```python
x = df['DIS'].values.reshape(-1,1)
y = df['MEDV'].values.reshape(-1, 1)
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
ransac.fit(x,y)
```

```python
#need to mark outliers

inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
np.arange(3,10,1)
```

```python
line_x = np.arange(3,10,1)
line_y = ransac.predict(line_x.reshape(-1,1))
```

```python
sns.set(style= 'darkgrid', context = 'notebook')
plt.figure(figsize = (8,6));
plt.scatter(x[inlier_mask], y[inlier_mask], c = 'pink', marker = 'o', label = 'Inliers')
plt.scatter(x[outlier_mask], y[outlier_mask], c = 'lightgreen', marker= 's', label= 'Outliers')

plt.plot(line_x, line_y, color = 'green')
plt.xlabel("average number of rooms per dwelling")
plt.ylabel("Median Value of Owner-occupied homes in $1000's")
plt.legend(loc = 'upper right')
plt.show()
```

### Standardise Variable to identify key Feature(S)

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(model.coef_, df.columns)), columns = ['coefficient', 'name']).set_index('name')
np.abs(result).sort_values(by = 'coefficient', ascending = False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
standard_coefficient_linear_reg = make_pipeline(scaler, model)
```

```python
standard_coefficient_linear_reg.fit(X,y)
result = pd.DataFrame(list(zip(standard_coefficient_linear_reg.steps[1][1].coef_, df.columns)), columns = ['coefficient', 'name']).set_index('name')
np.abs(result).sort_values(by = 'coefficient', ascending = False)
```

### Use R2 to identify Key Features

```python
from sklearn.metrics import r2_score
```
